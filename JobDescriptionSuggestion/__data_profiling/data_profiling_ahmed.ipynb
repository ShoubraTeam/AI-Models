## Basic Exploration

## Cleaning

**Addressing nulls & duplicates**

```python
df1.isna().sum()
```

```python
df1.duplicated().sum()
```

```python
df1.drop_duplicates(inplace = True)

df1.shape
```

**Clean Names**

```python
names_dict = {
    'positionName' : 'title',
    'searchInput/position' : 'category'
}
df1.rename(columns = names_dict, inplace = True)
df1.head()
```

```python
# add the year
df1['year'] = 2025
df1.head()
```

**The `description` column**

```python
def print_job_descriptions(descriptions, n_sep = 100):
    for idx, desc in enumerate(descriptions):
        print(f'>> Job #_{idx}')
        print()
        print(desc)
        print()
        print('-' * n_sep)
```

```python
# explore samples
descs = df1.description.values.tolist()
print_job_descriptions(descs[:5])
```

```python
def get_description_lengths(desc: str):
    words = desc.split()
    return len(words)
```

```python
df1['description_lengths'] = df1.description.apply(get_description_lengths)

df1.head()
```

```python
df1.description_lengths.describe()
```

```python
df1.sort_values(by = ['description_lengths'])['description_lengths'].head(n = 20)
```

```python
filt = df1['description_lengths'] <= 100
descs = df1[filt].description.values.tolist()
```

```python
print_job_descriptions(descriptions = descs)
```

```python
df1 = df1.copy().drop(columns = ['category', 'description_lengths'])
df1.head()
```
