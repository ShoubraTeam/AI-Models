**Text Files**

```python
TXT_ROOT_PATH = 'data/jobs_2_data/'
position_paths = [os.path.join(TXT_ROOT_PATH, position) for position in os.listdir(TXT_ROOT_PATH)]

print(position_paths)
```

```python
def load_txt_file(path):
    with open(path, 'r') as f:
        content = f.read()
    return content
```

```python
import re
def extract_data(job: str):
    job = job.split(']')[1].strip()
    lines = job.split("\n\n")
    return {
        'title' : lines[0].split(":")[1].strip(),
        'responsibilities' : lines[1].split(":")[1].strip().split(", "),
        'tech_skills' : lines[2].split(":")[1].strip().split(", "),
        'tools' : lines[3].split(":")[1].strip().split(", "),
        'soft_skills' : lines[4].split(":")[1].strip().split(", "),
    }
```

```python
def format_data(data: dict) -> str:
    """
    Turns the data dictionary into a formal job description.
    """    
    def to_bullets(items):
        cleaned_items = []
        for item in items:
            # Remove "and " if it's at the start of the string
            clean_item = item.strip()
            if clean_item.lower().startswith("and "):
                clean_item = clean_item[4:]
            
            # Capitalize the first letter and add bullet
            cleaned_items.append(f"• {clean_item.capitalize()}")
        
        return "\n".join(cleaned_items)

    job_desc = f"""
Key Responsibilities
{to_bullets(data.get('responsibilities', []))}

Technical Skills
{to_bullets(data.get('tech_skills', []))}

Soft Skills
{to_bullets(data.get('soft_skills', []))}

Tools & Software
{to_bullets(data.get('tools', []))}
"""
    return job_desc
```

```python
# reading the .txt jobs into csv
txt_data = []
for pos_idx, pos_folder in enumerate(position_paths, start = 1):
    position = pos_folder.split("/")[-1]
    print(f'>> Position #_{pos_idx} - {position}')

    # all files in that position folder
    n_error_files = 0
    for file_idx, file in enumerate(os.listdir(pos_folder)):
        try:
            print(end = "\t")
            print(f'Finished {file_idx} files')
            
            # skip `Fundamental skills & Recent Trends`
            if "Job" not in file:
                continue
            
            # read file
            file_path = os.path.join(pos_folder, file)
            file_content = load_txt_file(file_path)
            data = extract_data(file_content)
    
            # extract data
            title = data['title']
            description = format_data(data)
            year = 2026
    
            # save
            txt_data.append({
                'title' : title,
                'description' : description,
                'year' : year
            })

        # skip errornos files
        except:
            n_error_files += 1
            pass

    print(f"Number of Files with Errors: {n_error_files}")

    print()
    print()
```

```python
txt_to_df = pd.DataFrame(txt_data)
txt_to_df.head()
```

## Chunking Data

**Exploring Lengths**

```python
df_final = pd.read_csv('data/final_data.csv', index_col = 0)
df_final.head()
```

```python
df_final['description_lengths'] = df_final['description'].str.split().apply(len)
df_final.head()
```

```python
df_final['description_lengths'].describe()
```

**Remove Rows without Sufficient Info**

```python
def print_job_descs(job_descs: list):
    for job_desc in job_descs:
        print(job_desc)
        print(100 * '-')
```

```python
filt = df_final['description_lengths'] <= 30
df_final[filt].shape
```

```python
small_job_descs = df_final[filt].sort_values(by = ['description_lengths'], ascending = False)['description'].values.tolist()

print_job_descs(small_job_descs[:5])
```

```python
# removing rows with len(desc) <= 30 --> no enough info included
df_final = df_final[~filt]
df_final.shape
```

```python
df_final.description_lengths.describe()
```

**Chunking Rows with Long Descriptions**

```python
def get_fixed_size_chunks(text: str, chunk_size: int = 700, overlap: int = 100) -> list:
    words = text.split()
    chunks = []
    
    i = 0
    while i < len(words):
        end = i + chunk_size
        chunk = " ".join(words[i:end])
        chunks.append(chunk)
        
        i += (chunk_size - overlap)
        
        if i + overlap >= len(words):
            break
            
    return chunks
```

```python
filt = df_final['description_lengths'] > 700
data_above_700 = df_final[filt].copy()
data_above_700.shape
```

```python
data_above_700.description_lengths.describe()
```

```python
data_above_700['description'] = data_above_700['description'].apply(
    lambda x : get_fixed_size_chunks(text = x)
)

data_above_700_expanded = data_above_700.copy().explode('description', ignore_index = True)
data_above_700_expanded.shape
```

```python
data_below_700 = df_final[~filt]
data_below_700.shape
```

```python
data_below_700.description_lengths.describe()
```

```python
df_final_v2 = pd.concat(
    objs = [data_below_700, data_above_700_expanded],
    axis = 0,
    ignore_index = True
).drop(columns = ['description_lengths'])

df_final_v2.shape
```

```python
df_final_v2.head()
```

```python
df_final_v2['description_lengths'] = df_final_v2['description'].str.split().apply(len)
df_final_v2.description_lengths.describe()
```
