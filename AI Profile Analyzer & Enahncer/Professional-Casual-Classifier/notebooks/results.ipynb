```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"..\models\baseline_cnn_logs.csv")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(df['epoch'], df['train_acc'], label='Train Acc', marker='o')
plt.plot(df['epoch'], df['val_acc'], label='Val Acc', marker='o')
plt.title('Training & Validation Accuracy')
plt.legend()

# رسم الـ Loss
plt.subplot(1, 2, 2)
plt.plot(df['epoch'], df['train_loss'], label='Train Loss', marker='s')
plt.plot(df['epoch'], df['val_loss'], label='Val Loss', marker='s')
plt.title('Training & Validation Loss')
plt.legend()

plt.show()
```
